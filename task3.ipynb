```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-20,20,1.2)
y = np.arange(-20,20,1.2)
X,Y = np.meshgrid(x,y)
xd = X/((X**2 + Y**2)**1.5)
yd = Y/((X**2 + Y**2)**1.5)
plt.quiver(X,Y,xd,yd, color = 'r')
plt.title("Electric field surrounding charge at (0,0)")
```

```python
# Magnetic field

x,y = np.meshgrid(np.arange(-20,20,1.2),np.arange(-20,20,1.2)) # coordinates for the position vector r
m = [0, 1]  # magnetic moment of the magnetic dipole moment

mr = m[0] * x + m[1] * y # dot product m * r

r_squared = x**2 + y**2
factor = 1 / ( np.sqrt(r_squared)**5 * 4 * np.pi) # pre factor, that is  multiplied with the vector 

u = factor * (3 * x * mr - m[0] * r_squared)
v = factor * (3 * y * mr - m[1] * r_squared)
t = xd *u
s = yd *v


plt.streamplot(x,y,t,s)
plt.title("Magnetic field")
plt.show()
print("Done")
```
